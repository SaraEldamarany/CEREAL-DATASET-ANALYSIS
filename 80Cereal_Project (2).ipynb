Exploratory Data Analysis (EDA) :

```python
df = df.sort_values(['rating'], ascending=False).reset_index(drop=True)



pink_palette = ["#FFB6C1", "#FFC0CB", "#FFD1DC", "#FFA6C9", "#7EA9B6"]

plt.figure(figsize=(12, 14))
sns.barplot(x=df["rating"], y=df["name"], palette = pink_palette)

for i in range(len(df)):
    plt.text(df["rating"][i] + 0.1, i, f'{df["rating"][i]:.2f}', va='center', fontsize=10, color="#FFA6C9" )


plt.xlabel("Rating", fontsize=15)
plt.ylabel("Product Name", fontsize=15)
plt.title("Product Rating for All Products", fontsize=22)
plt.show()
```

```python
import plotly.express as px

df = df.sort_values(['rating'], ascending=False).reset_index(drop=True)
df['rating_category'] = pd.cut(df['rating'], bins=5, labels=["Very Low", "Low", "Medium", "High", "Very High"])

fig = px.bar(
    df,
    x="rating",
    y="name",
    orientation="h",
    color="rating_category",
    color_discrete_sequence=pink_palette
)

fig.update_layout(
    height=600,
    title="Product Rating for All Products",
    xaxis_title="Rating",
    yaxis_title="Product Name",
    yaxis=dict(autorange="reversed")
)

fig.show()
```

```python
# Define nutritional value criteria (high protein, fiber, vitamins, low sugar)
df['nutritional_value'] = df['protein'] + df['fiber'] + (df['vitamins'] / 100) - df['sugars']

# Sort cereals by highest nutritional value
top_nutritional_cereals = df.sort_values(by='nutritional_value', ascending=False)[['name', 'nutritional_value']]
top_nutritional_cereals.head(10).plot(kind='barh', x='name', y='nutritional_value', title="Top 10 Cereals by Nutritional Value", color='#F6BFC7', figsize=(12, 5))
```

```python
# Grouped bar plot: Nutritional content by Manufacturer
nutritional_factors = ['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']
manufacturer_labels = ['American Home Food Products', 'General Mills', 'Kelloggs', 'Nabisco', 'Post', 'Quaker Oats', 'Ralston Purina']

# Calculate the mean values for each nutritional factor and manufacturer
mean_values = df.groupby('mfr')[nutritional_factors].mean()
#pink_palette2 = ["#FFA6C9", "#7EA9B6","#FFB6C1", "#FFC0CB", "#FFD1DC"]
#custom_palette = ["#F6BFC7", "#B8D0D7"]
pink_palette2= ["#D95A85","#4A4A4A", "#50778A","#FFB6C1", "#7A7A7A",  "#5A8EC3", "#D16E82","#FFA6C9", "#C9708A","#FFD1DC", "#BA6C78"]

# Create the grouped bar plot
plt.figure(figsize=(12, 6))
ax = mean_values.plot(kind='bar', color=pink_palette2[:len(manufacturer_labels)])
plt.xlabel('Manufacturer')
plt.ylabel('Mean Value')
plt.title('Nutritional Content by Manufacturer')
ax.legend(nutritional_factors, loc='center left', bbox_to_anchor=(1, 0.5))  # Place legend to the right
plt.xticks(range(len(manufacturer_labels)), manufacturer_labels, rotation='horizontal')
plt.xticks(rotation=45, ha='right')
plt.show()
```
